### 🔄 3. Data Preprocessing
---

```python
# Import libraries untuk preprocessing
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 🔍 a ) Data Cleaning
---

```python
# Data Cleaning
df = df.dropna()                         # Hapus nilai kosong
df = df.drop_duplicates('nama_kos')      # Hapus duplikat

df = df.drop(columns=['nama_kos', 'url', 'alamat', 'scraped_at'])
df = df.fillna(0)
```

```python
df.info()
```

```python
df.shape
```

#### ⚛️ b ) Data Transformation
---

```python
# Type conversion
df['harga'] = df['harga'].astype(str).str.replace(r'[^\d]', '', regex=True).astype(float)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['jumlah_review'] = df['jumlah_review'].astype(str).replace('', '0').str.replace(r'[^\d]', '', regex=True).astype(float)
df['total_transaksi'] = df['total_transaksi'].astype(str).replace('', '0').str.replace(r'[^\d]', '', regex=True).astype(float)
```

```python
# Encode 
df['jenis_kos_encoded'] = df['jenis_kos'].astype('category').cat.codes
df['area_encoded'] = df['area'].astype('category').cat.codes
df['region_encoded'] = df['region'].astype('category').cat.codes
```

```python
# Scaling
num_cols = df.select_dtypes(include=['float64', 'int64']).columns

scaler = MinMaxScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
# Normalization / Log Transform
df['harga_log'] = np.log1p(df['harga'])
df['jumlah_review_log'] = np.log1p(df['jumlah_review'])
df['total_transaksi_log'] = np.log1p(df['total_transaksi'])
```

```python
# Discretization
df['harga_bin'] = pd.cut(df['harga'], bins=5, labels=False)
df['rating_bin'] = pd.cut(df['rating'], bins=5, labels=False)
df['jumlah_review_bin'] = pd.cut(df['jumlah_review'], bins=5, labels=False)
df['total_transaksi_bin'] = pd.cut(df['total_transaksi'], bins=5, labels=False)
```

```python
df.shape
```

#### 🛠️ c ) Special Data Transformation

```python
df['region'].unique()
```

```python
# Special Encode untuk region -> kota
def extract_kota(region: str):
    region = str(region).lower()
    parts = region.split('-')
    
    # Cari kata "kota"
    if "kota" in parts:
        idx = parts.index("kota")
        if idx + 1 < len(parts):
            return parts[idx + 1].capitalize()
    
    # Kalau tidak ada "kota", ambil bagian pertama (nama kota/kabupaten)
    return parts[0].capitalize()

df['kota'] = df['region'].apply(extract_kota)

df = pd.get_dummies(df, columns=['jenis_kos', 'area', 'kota'])
df = df.drop(columns=['region'])
```

```python
# List all one-hot encoded kota columns
kota_cols = [col for col in df.columns if col.startswith('kota_')]
print(kota_cols)
```

```python
df['periode'].unique()
```

```python
# Lihat jumlah data tiap kategori periode
print(df['periode'].value_counts())

# Lihat contoh data dari masing-masing kategori
print(df[df['periode'] == '/bulan'].head(15).harga)
print(df[df['periode'] == '(Bulan pertama)'].head(15).harga)
```

```python
# Special Encode untuk periode
def encode_periode(x):
    if '/bulan' in str(x).lower():
        return 1
    elif 'bulan pertama' in str(x).lower():
        return 2
    else:
        return 0
df['periode_encoded'] = df['periode'].apply(encode_periode)
```

```python
# Special Feature Creation untuk landmarks
df['num_landmarks'] = df['landmarks'].apply(lambda x: len(x) if isinstance(x, list) else 0)
def min_jarak_landmarks(x):
    if isinstance(x, list):
        jarak_list = []
        for d in x:
            if isinstance(d, dict) and 'jarak' in d:
                # Ambil angka dari string jarak, misal '1.1 km' -> 1.1
                try:
                    jarak = float(str(d['jarak']).split()[0].replace(',', '.'))
                    jarak_list.append(jarak)
                except ValueError:
                    continue
        return min(jarak_list) if jarak_list else None
    return None
df['min_jarak_landmark'] = df['landmarks'].apply(min_jarak_landmarks)
```

```python
# Special Feature Creation untuk fasilitas.ukuran_listrik
def extract_ukuran_listrik(x):
    ukuran = None
    listrik = 0
    if isinstance(x, list):
        for item in x:
            if 'meter' in item:
                try:
                    ukuran = float(item.split('x')[0].strip())
                except ValueError:
                    ukuran = None
            if 'termasuk listrik' in item.lower():
                listrik = 1
    return pd.Series({'ukuran_kamar': ukuran, 'listrik_included': listrik})

df[['ukuran_kamar', 'listrik_included']] = df['fasilitas.ukuran_listrik'].apply(extract_ukuran_listrik)
```

#### 🤖 d ) Feature Engineering
---

```python
# Feature Engineering - Count features
df['num_peraturan'] = df['peraturan'].apply(lambda x: len(x) if isinstance(x, list) else 0)
df['num_landmarks'] = df['landmarks'].apply(lambda x: len(x) if isinstance(x, list) else 0)
df['num_fasilitas_kamar'] = df['fasilitas.kamar'].apply(lambda x: len(x) if isinstance(x, list) else 0)
df['num_fasilitas_kamar_mandi'] = df['fasilitas.kamar_mandi'].apply(lambda x: len(x) if isinstance(x, list) else 0)
df['num_fasilitas_umum'] = df['fasilitas.umum'].apply(lambda x: len(x) if isinstance(x, list) else 0)
df['num_fasilitas_parkir'] = df['fasilitas.parkir'].apply(lambda x: len(x) if isinstance(x, list) else 0)
```

```python
df.shape
```

#### 📉 e ) Outlier Handling
---

```python
# Outlier Handling - IQR dan Z-score
for col in ['harga', 'rating', 'jumlah_review']:
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    sns.histplot(df[col], bins=50, kde=True)
    plt.title(f'Distribusi {col}')
    plt.subplot(1,2,2)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot {col}')
    plt.show()

for col in ['harga', 'rating', 'jumlah_review']:
    print(f"Skewness {col}: {df[col].skew():.2f}")
```

```python
# IQR Filter
def iqr_filter(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    return df[(df[col] >= Q1 - 1.5 * IQR) & (df[col] <= Q3 + 1.5 * IQR)]

df = iqr_filter(df, 'harga')
```

```python
df.shape
```

#### 🔁 f ) Re-Cleaning (Akhir Preprocess)
---

```python
df = df.dropna(subset= ['harga'])
df = df.drop(columns=[
    'peraturan',
    'landmarks',
    'fasilitas.kamar',
    'fasilitas.kamar_mandi',
    'fasilitas.umum',
    'fasilitas.parkir',
    'periode', 
    'fasilitas.ukuran_listrik'
])


df['rating'] = df['rating'].fillna(0)
df['jumlah_review'] = df['jumlah_review'].fillna(0)
df['rating_bin'] = df['rating_bin'].fillna(0)
df['min_jarak_landmark'] = df['min_jarak_landmark'].fillna(df['min_jarak_landmark'].max())  # misal isi dengan max jarak
df['ukuran_kamar'] = df['ukuran_kamar'].fillna(0)
```
